B. Write a function that takes one press release string as an input and:

- Removes named entities from each press release string (**Hint**: you may want to use `re.sub` with an or condition)
- Scores the sentiment of the entire press release using the `SentimentIntensityAnalyzer` and `polarity_scores`
- Returns the length-four (negative, positive, neutral, compound) sentiment dictionary (any order is fine)

Apply that function to each of the press releases in `doj_subset`. 

**Hints**: 

- A function + list comprehension to execute will takes about 30 seconds on a respectable local machine and about 2 mins on jhub; if it's taking a very long time, you may want to check your code for inefficiencies.
```python
## your code here to define function
def remove_named_entities_and_score_sentiment(press_release):
    named_entity_pattern = r'\b[A-Z][a-z]*\b(?:\s[A-Z][a-z]*\b)*'
    clean_press_release = re.sub(named_entity_pattern, '', press_release)
    analyzer = SentimentIntensityAnalyzer()
    
    sentiment_scores = analyzer.polarity_scores(clean_press_release)
    
    # Create and return the length-four sentiment dictionary
    length_four_sentiment = {
        'negative': sentiment_scores['neg'],
        'positive': sentiment_scores['pos'],
        'neutral': sentiment_scores['neu'],
        'compound': sentiment_scores['compound']
    }
    
    return clean_press_release, length_four_sentiment
```

```python
def process_press_release(press_release):
    clean_press_release, sentiment_scores = remove_named_entities_and_score_sentiment(press_release)
    return clean_press_release, sentiment_scores

doj_subset_processed = doj_subset['contents'].apply(process_press_release)
doj_subset_processed_df = pd.DataFrame(doj_subset_processed.tolist(), columns=['Clean_Press_Release', 'Sentiment_Scores'])

# Concatenate the new DataFrame with the original doj_subset DataFrame
doj_subset_with_sentiment = pd.concat([doj_subset.reset_index(drop=True), doj_subset_processed_df], axis=1)
doj_subset_with_sentiment.head()
```

For steps B - D, to receive full credit, write a function `get_topwords` that helps you avoid duplicated code when you find top words for the different subsets of the data. There are different ways to structure it but one way is to feed it subsetted data (so data subsetted to one topic etc.) and for it to get the top words for that subset.


```python

def create_dtm(list_of_strings, metadata):
    vectorizer = CountVectorizer(lowercase = True)
    dtm_sparse = vectorizer.fit_transform(list_of_strings)
    dtm_dense_named = pd.DataFrame(dtm_sparse.todense(), 
        columns=vectorizer.get_feature_names_out())
    dtm_dense_named_withid = pd.concat([metadata.reset_index(), dtm_dense_named], axis = 1)
    return(dtm_dense_named_withid)
```

```python
dtm_doj = create_dtm(list_of_strings=doj_subset_wscore['processed_text'],
                     metadata=doj_subset_wscore[['id', 'compound', 'topics_clean']])


def get_topwords(dtm, num_words=10):
    word_counts = dtm[dtm.columns[4:]].sum(axis=0).sort_values(ascending=False)
    return word_counts.head(num_words)

# thresholds for top and bottom 5% compound scores
top_5_percent_threshold = dtm_doj['compound'].quantile(0.95)
bottom_5_percent_threshold = dtm_doj['compound'].quantile(0.05)

# get subsets based on thresholds
top_5_percent_dtm = dtm_doj[dtm_doj['compound'] >= top_5_percent_threshold]
bottom_5_percent_dtm = dtm_doj[dtm_doj['compound'] <= bottom_5_percent_threshold]

print("top 10 words in the top 5% most positive:")
print(get_topwords(top_5_percent_dtm))
print("top 10 words in the bottom 5% most negative:")
print(get_topwords(bottom_5_percent_dtm))

# for topics
for topic in dtm_doj['topics_clean'].unique():
    topic_dtm = dtm_doj[dtm_doj['topics_clean'] == topic]
    print("top 10 words for topic " + topic + ":")
    print(get_topwords(topic_dtm))
```
